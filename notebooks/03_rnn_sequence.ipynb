# Deep Learning Tutorial - RNN 序列演示

使用 `SimpleRNN` 处理简短序列，查看隐藏状态与输出。

## 导入与样例序列
- 输入为长度 3 的标量序列
- 输出维度 2，便于观察 softmax 分布

```python
from deep_learning.architectures import SimpleRNN

sequence = [[1.0], [0.0], [1.0]]
print("样例序列:", sequence)
```

## RNN 前向
- 隐藏维度 2，输出维度 2
- 展示最后一步输出与隐藏状态

```python
rnn = SimpleRNN(input_size=1, hidden_size=2, output_size=2, learning_rate=0.05)
outputs, hidden_states = rnn.forward(sequence)

print("最后一步输出分布:", outputs[-1])
print("最终隐藏状态:", hidden_states[-1])
```
