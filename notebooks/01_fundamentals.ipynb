## 感知机 (AND 门)
- 训练一个二分类感知机实现 AND 逻辑。

```python
X_and = [[0, 0], [0, 1], [1, 0], [1, 1]]
y_and = [0, 0, 0, 1]

perceptron = Perceptron(input_size=2, learning_rate=0.2)
perceptron.train(X_and, y_and, epochs=8)

preds = [perceptron.predict(x) for x in X_and]
print("预测: ", preds)
assert preds == y_and
print("AND 门学习成功 ✅")
```

## MLP 前向示例
- 构建一个 2-4-1 的网络，演示前向输出。

```python
mlp = MLP(layers=[2, 4, 1], learning_rate=0.01)

sample = [0.5, -0.1]
activations, z_values = mlp.forward(sample, training=False)
output = activations[-1]

print("输入: ", sample)
print("输出: ", output)
print("前向传播完成 ✅")
```
